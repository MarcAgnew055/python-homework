```python
import os

import csv

mypath = '/Users/marcagnew/GitHub/python-homework/PyBank'

budgetCSV = os.path.join(mypath, 'budget_data.csv')

with open(budgetCSV, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')

    header = next(csvreader)

    rows = 0
    total = 0
    Profit = 0
    Loss = 0
    row_profit = 0
    row_loss = 0
    Max = 0
    Min = 0

    last_ProfLost = 0
    Total_months_list = []
    ProfLost_change = 0
    Month_change = 0
    Total_ProfLost_list = []
    current_ProfLost = 0
    ProfLost_change_list = []

    # Loop 

    for row in csvreader:
        rows += 1
        total += int(row[1])

        if (int(row[1]) > 0):
            Profit += int(row[1])
            row_profit += 1
        else:
            Loss += int(row[1])
            row_loss += 1

        # Average in "Profit/Losses" 

        Total_months_list.append(row[0])
        Total_ProfLost_list.append(int(row[1]))

            # Greatest increase and decrease in profits 

        if (int(row[1]) > Max):
            Max = int(row[1])
            Month_max = row[0]
        elif (int(row[1]) < Min):
            Min = int(row[1])
            Month_min = row[0]

    for i in range(len(Total_ProfLost_list) - 1):
        ProfLost_change_list.append(Total_ProfLost_list[i + 1] - Total_ProfLost_list[i])

    Average_change = (sum(ProfLost_change_list)) / (len(ProfLost_change_list))


# Print the analysis
print("Financial Analysis")
print("-----------------------\n")
print(f"Total Months: {str(rows)} \n")
print(f"Total: ${str(total)}\n")
print(f"Average Change: ${str('%.2f' % Average_change)}\n")
print(f"Greatest Increase in Profits: {Month_max} (${str(Max)})\n")
print(f"Greatest Decrease in Profits: {Month_min} (${str(Min)})\n")

# Export a text file with the results
results_file = ('/Users/marcagnew/GitHub/python-homework/PyBank/Results.txt')

# Save the output file path
with open(results_file, 'w') as file:
    file.write("Financial Analysis")
    file.write("-----------------------\n")
    file.write(f"Total Months: {str(rows)} \n")
    file.write(f"Total: ${str(total)}\n")
    file.write(f"Average Change: ${str('%.2f' % Average_change)}\n")
    file.write(f"Greatest Increase in Profits: {Month_max} (${str(Max)})\n")
    file.write(f"Greatest Decrease in Profits: {Month_min} (${str(Min)})\n")
```
